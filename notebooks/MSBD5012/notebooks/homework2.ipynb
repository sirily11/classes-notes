```
w1 = np.array([[1, -1], [-1, 1]])
w2 = np.array([[-1, -1], [1, 1]])
w3 = np.array([[1], [1]])
x = np.array([1, 2])
```

```
import math
from numpy import tanh

def tanh_derivative(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return dt

def sigmoid(X):
   return 1/(1+np.exp(-X))


def sigmoid_derivative(x):
    outcome = sigmoid(x)
    return outcome * (1 - outcome)
```

```
layer1 = x.dot(w1)
layer1 = tanh(layer1)
layer1
```

```
layer2 = layer1.dot(w2)
print(layer2)
layer2 = tanh(layer2)
layer2
```

```
output = layer2.dot(w3)
output *2
```

```
e = sigmoid(output) 
e
```

backprop

$$error_{hidden} = w^T * error_{output}$$

```
b_layer2 = e * tanh_derivative(layer2)
b_layer2
```

```
b_layer1_err = b_layer2.dot(w2.T)
b_layer1 = b_layer1_err * tanh_derivative(layer1)
b_layer1
```
