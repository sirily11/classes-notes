```
def output_cnn(w1, f, p, s):
    """
    w1: input size
    f: filter size
    p: number of zero padding
    s: the stride
    k: number of filters
    """
    return (w1 - f + 2 * p) / s + 1
```

```
def number_of_parameter(f, d, k):
    return (f * f *d + 1) * k
```

```
def number_of_flops(f, d, w2, h2, d2):
    return (f * f * d + 1) * w2 * h2 * d2
```
